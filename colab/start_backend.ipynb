```python
# 1) リポジトリを取得
!git clone https://github.com/TaiyoYamada/voice-comic-theater.git
%cd /content/voice-comic-theater
```

```python
# 2) 設定（秘密情報は Colab の「シークレット」から読み込む。直書きしない）
import os
from google.colab import userdata  # 左の鍵アイコンから GAS_URL（と必要なら NGROK_AUTHTOKEN）を登録

os.environ['GAS_URL'] = userdata.get('GAS_URL')

# 公開トンネル: 本番は 'cloudflare'（無料・鍵不要・複数台OK）。手元のngrokを使うなら 'ngrok'。
os.environ['TUNNEL'] = 'cloudflare'

# ngrok を使う場合だけ NGROK_AUTHTOKEN が必要（cloudflare なら不要）。
if os.environ['TUNNEL'] == 'ngrok':
    os.environ['NGROK_AUTHTOKEN'] = userdata.get('NGROK_AUTHTOKEN')

# この台の識別情報（台ごとに変える）
os.environ['SERVER_ID']    = 'colab-1'
os.environ['SERVER_COLOR'] = 'red'
os.environ['SERVER_LABEL'] = '赤サーバー'
os.environ['CAPACITY']     = '2'  # 1台あたり1〜2人
```

```python
# 3) 起動（依存インストール → FastAPI起動 → トンネル公開 → GAS登録 → heartbeat）
#    このセルは実行したままにしておく（止めるとサーバーが落ちる）。
%run colab/colab_runner.py
```
